```python
import requests
from bs4 import BeautifulSoup
```

```python
## generic funciton to just get complete html in soup object
def get_html(url):
    response = requests.get(url)
    return BeautifulSoup(response.text, 'html.parser')

```

```python
## here getting the main main categories names here from olx home page

def get_links(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    links = soup.select("#body-wrapper > div._948d9e0a._95d4067f._1e086faf._4122130d > div._2d92de1e > div > div._19d47bcd > div:nth-child(2) a")
    category_links = []
    for link in links:
        href = link.get('href')
        text = link.get_text(strip=True)
        category_links.append({
            'text': text,
            'href': url+href
        })
    return category_links
```

```python
### getting inside each link in the dictionary of main links
def get_inside_links(category_links):
    inside_links = []
    data = []
    for category in category_links:
        soup = get_html(category['href'])
        container = soup.select_one("#body-wrapper > div._948d9e0a._95d4067f._1e086faf._4122130d > header:nth-child(4) > div > div > div > div._948d9e0a._53c2e845._95d4067f.e1c7c3d4._4122130d > div._948d9e0a.cbff116c.e1c7c3d4 > div._14492788 > div > ul")
        list_items = container.find_all('li')
        filtered_items = []
        for item in list_items:
            if item.get('aria-label') == 'Listing':
                filtered_items.append(item)
        list_item=[]
        list_item=filtered_items;
        for eachli in list_item:
            link = eachli.find('a')
            href = link.get('href')
            href = "https://www.olx.com.pk"+href
            product_data=create_table_for_each_product(href)
            data.append(product_data)
            make_excel(data)
            print(product_data)




```
